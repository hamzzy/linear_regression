## Predicting weight using linear regression

### Import the height-weight data csv

```python
import numpy as np
import pandas as pd
filename = "height-weight.csv"

data =pd.read_csv(filename) 
data=np.array(data)
data
```

```python
# We check the size of the data
dat=pd.DataFrame(columns=["height","weight"],data=data)
dat
```

```python
# We check the first 10 rows
dat[:10]
```

```python
#We import the plotting library 
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
```

```python
#We separate out the independent variable height into X 
#and dependent variable weight into y
X=dat["height"]
y=dat["weight"]
X
```

```python
X_20 = X[:10]
y_20 = y[:10]
plt.scatter(X_20, y_20, color='red', s=10) ## here s is size
```

```python
from sklearn.model_selection import train_test_split
# Split the data into training/testing sets
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1)

```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# Modeify the data to input into sklearn
X_train=np.array(X_train)
X_train=np.reshape(X_train,(-1,1))
X_test=np.array(X_test)
X_test=np.reshape(X_test,(-1,1))
```

```python
X_test
```

```python
# Create linear regression object
regr = linear_model.LinearRegression()
```

```python
# Train the model using the training sets
regr.fit(X_train, y_train)
```

```python

# Make predictions using the testing set
y_pred = regr.predict(X_test)
```

```python
# The coefficients
print('Coefficients: \n', regr.coef_)
```

```python
# Plot outputs
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
